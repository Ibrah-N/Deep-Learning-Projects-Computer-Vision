## Tensors Initialization

```python
import torch
```

```python
# empty tensor


t_1 = torch.empty(size=(3, 3))
t_1
```

```python
# check tensor type

print(type(t_1))
print(t_1.dtype)
```

```python
# zero tensor

t_2 = torch.zeros(size=(3, 3))
t_2
```

```python
# rand tensor

t_3 = torch.rand(3, 3)
t_4 = torch.randn(4, 4)
t_5 = torch.randint(size=(2, 2), low=0, high=5)

print(t_3)
print(t_4)
print(t_5)
```

```python
# tensor from tensor

t_6 = torch.tensor([
    [1, 2],
    [2, 2]
])

print(t_6, t_6.shape)
```

```python
# tensor of linespace

t_7 = torch.linspace(start=1, end=5, steps=5)

print(t_7)
```

```python
# seed use

torch.manual_seed(1)
t_8 = torch.rand(size=(2, 2))

print(t_8)
```

```python
torch.manual_seed(1)
t_9 = torch.rand(size=(2, 2))

print(t_9)
```

```python
# eye; identity tensors


t_10 = torch.eye(n=10, m=3)
print(t_10)
```

```python
# like function

t_11 = torch.ones_like(t_10)
print(t_11)
```

## Tensors types

```python
# initialize tensor
t_01 = torch.randint(size=(3, 3), low=2, high=20)
t_03 = torch.randint(size=(3, 3), low=2, high=20, dtype=torch.float32)


print(f"tensor: \n{t_01}, \nshape: {t_01.shape}, type={t_01.dtype}")
print(f"tensor: \n{t_03}, \nshape: {t_03.shape}, type={t_03.dtype}")
```

```python
# cast the type into float

t_02 = t_01.to(torch.float32)

print(f"tensor: \n{t_02}, \nshape: {t_02.shape}, type={t_02.dtype}")
```

## Mathematical Operations

### Scalar Operations

```python
# initialization

t_20 = torch.randint(size=(3, 3), low=2, high=20)
t_21 = torch.randint(size=(3, 3), low=2, high=40)
t_22 = torch.randint(size=(4, 4), low=20, high=80)

print(f"tensor: \n{t_20}, \nshape: {t_20.shape}")
print(f"tensor: \n{t_21}, \nshape: {t_21.shape}")
print(f"tensor: \n{t_22}, \nshape: {t_22.shape}")
```

```python
# addition

t_23 = t_20 + t_21


print(f"{t_20} \n+ \n{t_21}   \n\n{t_23}")
```

```python
# subtraction

t_23 = t_20 - t_21


print(f"{t_20} \n- \n{t_21}   \n\n{t_23}")
```

```python
print(torch.relu(t_23))
```

```python
# subtraction

t_23_sub = t_20 - t_21


print(f"{t_20} \n- \n{t_21}   \n\n{t_23_sub}")
print(f"\nApplying ReLU to t_23_sub:\n{torch.relu(t_23_sub)}")
```

```python
# multiplication

t_23 = t_20 * t_21


print(f"{t_20} \n* \n{t_21}   \n\n{t_23}")
```

```python
# division

t_23 = t_20 / t_21


print(f"{t_20} \n/ \n{t_21}   \n\n{t_23}")
```

```python
# power

t_23 = t_20 / t_21


print(f"{t_20} \n* \n{t_21}   \n\n{t_23}")
```
